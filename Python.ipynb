```python
import gdown

file_id = '1u2LOhVHSFIW5IMxTxhTQ7BX-Bvu1HfzR'
file_path = 'canada_by_citizenship.csv'
gdown.download(f'https://drive.google.com/uc?id={file_id}', file_path, quiet=True)

print(f'File downloaded to {file_path}')
```

```python
#3
df_can = pd.read_csv(
    file_path,
    skipfooter=2,
    engine='python' # Specify python engine due to skipfooter
)
print('Data downloaded and read into a dataframe!')
```

```python
df_can["Date"] = pd.to_datetime(df_can["Date"], format="%m-%d-%y", errors="coerce")
```

```python
df_can= df_can.drop(columns=["index", "Unnamed: 22"], errors="ignore")
```

```python
df_can = df_can.dropna(subset=["Amount", "currency"])
```

```python
sales_by_category = df_can[df_can["Status"].str.contains("Shipped", na=False)].groupby("Category")["Amount"].sum().sort_values(ascending=False)
display(sales_by_category)
```

```python
top_sku = df_can[df_can["Status"].str.contains("Shipped", na=False)].groupby("SKU")["Amount"].sum().nlargest(10)
display(top_sku)
```

```python
city_sales = df_can[df_can["Status"].str.contains("Shipped", na=False)].groupby("ship-city")["Amount"].sum().nlargest(10)
display(city_sales)
```

```python
import matplotlib.pyplot as plt

sales_by_category.plot(kind='bar', figsize=(10, 5))
plt.title("Total Penjualan per Kategori")
plt.ylabel("Total Amount (INR)")
plt.xlabel("Kategori")
plt.xticks(rotation=-45, ha='left') # Rotate x-axis labels by -45 degrees, align left
plt.tight_layout()
plt.show()
```

```python
daily_sales = df_can[df_can["Status"].str.contains("Shipped", na=False)].groupby("Date")["Amount"].sum()

daily_sales.plot(figsize=(12, 6))
plt.title("Tren Penjualan Harian")
plt.ylabel("Total Sales")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```
